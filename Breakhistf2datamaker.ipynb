```python
#  Breakhistf2datamaker.ipynb  v1.0.0 
# 
#  --------------------------------------------------
#  Hangzhou Domain Zones Technology Co., Ltd

#  Apache Licence 2.0       https://www.apache.org/licenses/LICENSE-2.0
#  --------------------------------------------------
 

import os
from shutil import copyfile 
   
directory_b = 'tf2breakhis'
    
if not os.path.exists(directory_b):
    os.makedirs(directory_b)
    
        
 
breakpath = '/Users/huanglican/yvsouBreakHis/BreaKHis_v1/BreaKHis_v1/histology_slides/breast/benign/SOB' 
benign_types = os.listdir( breakpath)
for benign_type in benign_types:
    directory_b8=  directory_b +'/'+benign_type
    if not os.path.exists(directory_b8):
        os.makedirs(directory_b8)
# Store the types. Example: 'adenosis', 'fibroadenoma', 'phyllodes_tumor', 'tubular_adenoma'.
benign_types = [x for x in benign_types if x!= '.DS_Store']

zooms = ['40', '100', '200', '400']

# Copy files from benign to other folder
for benign_type in benign_types:
    #patients = os.listdir('/breast/benign/SOB/'+benign_type)
    patients = os.listdir(breakpath +'/'+ benign_type  )
    
    patients = [x for x in patients if x!= '.DS_Store']
    
    for patient in patients:
        for zoom in zooms:
            images = os.listdir(breakpath+'/'+benign_type+'/'+patient+'/'+zoom+'X')
            images = [x for x in images if x!= '.DS_Store']
            for image in images:
                copyfile(breakpath+'/'+benign_type+'/'+patient+'/'+zoom+'X'+'/'+image, directory_b +'/'+benign_type+'/breast_'+zoom+'_benign_'+image)

breakmalignantpath = '/Users/huanglican/yvsouBreakHis/BreaKHis_v1/BreaKHis_v1/histology_slides/breast/malignant/SOB' 
malignant_types = os.listdir( breakmalignantpath)

for malignant_type in malignant_types:
    directory_m8= directory_b +'/' +malignant_type
    if not os.path.exists(directory_m8):
        os.makedirs(directory_m8)
# Store the types. Example: 'adenosis', 'fibroadenoma', 'phyllodes_tumor', 'tubular_adenoma'.
malignant_types = [x for x in malignant_types if x!= '.DS_Store']

zooms = ['40', '100', '200', '400']

# Copy files from benign to other folder
for malignant_type in malignant_types:
    #patients = os.listdir('/breast/benign/SOB/'+benign_type)
    patients = os.listdir(breakmalignantpath +'/'+ malignant_type ) 
    
    patients = [x for x in patients if x!= '.DS_Store']
    
    for patient in patients:
        for zoom in zooms:
            images = os.listdir(breakmalignantpath+'/'+malignant_type+'/'+patient+'/'+zoom+'X')
            images = [x for x in images if x!= '.DS_Store']
            for image in images:
                copyfile(breakmalignantpath+'/'+malignant_type+'/'+patient+'/'+zoom+'X'+'/'+image, directory_b +'/'+malignant_type+'/breast_'+zoom+'_malignant_'+image)



```
